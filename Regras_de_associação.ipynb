## Base de dados mercado 1

```python
base_mercado1 = pd.read_csv('mercado.csv', header=None)
base_mercado1
```

```python
base_mercado1.shape
```

```python
transacoes = []
for i in range(len(base_mercado1)):
    transacoes.append([str(base_mercado1.values[i,j]) for j in range(base_mercado1.shape[1])])
transacoes
```

```python
type(transacoes)
```

```python
regras = apriori(transacoes, min_support = 0.3, min_confidence = 0.8, min_lift=2)
resultados = list(regras)
resultados
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
    s = resultado[1]
    result_rules = resultado[2]
    for result_rule in result_rules:
        a = list(result_rule[0])
        b = list(result_rule[1])
        c = result_rule[2]
        l = result_rule[3]
        A.append(a)
        B.append(b)
        suporte.append(s)
        confianca.append(c)
        lift.append(l)
```

## Base de dados mercado 2

```python
base_mercado2 = pd.read_csv('mercado2.csv', header = None)
base_mercado2
```

```python
base_mercado2.shape
```

```python
len(base_mercado2) == base_mercado2.shape[0]
```

```python
transacoes = []
for i in range(len(base_mercado2)):
    transacoes.append([str(base_mercado2.values[i, j]) for j in range(base_mercado2.shape[1])])
transacoes
```

Produtos que são vendidos 4 vezes por dia

```python
intencao = 4
```

```python
vendas = intencao * 7 # A base contém as vendas feitas em uma semana (7 dias)
vendas
```

```python
suporte = vendas / len(base_mercado2) # Quantidade de transações feitas na semana (7501)
suporte
```

```python
regras = apriori(transacoes, min_support=suporte, min_confidence=0.2, min_lift=3)
resultados = list(regras)
resultados
```

```python
A = []
B = []
suporte = []
confianca = []
lift = []

for resultado in resultados:
    s = resultado[1]
    result_rules = resultado[2]
    for result_rule in result_rules:
        a = list(result_rule[0])
        b = list(result_rule[1])
        c = result_rule[2]
        l = result_rule[3]
        A.append(a)
        B.append(b)
        suporte.append(s)
        confianca.append(c)
        lift.append(l)
```

```python
rules_df = pd.DataFrame({'A': A, 'B': B, 'Suporte': suporte, 'Confiança': confianca, 'Lift': lift})
rules_df
```

```python
rules_df.sort_values(by='Suporte', ascending=False)
```
